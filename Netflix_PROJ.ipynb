Let's start with step 1: Loading the dataset.

```python
import pandas as pd

# Load the dataset
file_path = '/Users/divyeshbhatt/Downloads/Datas/SQL/netflix_data.csv'  # Replace with your file path
netflix_data = pd.read_csv(file_path)
```

```python
# Filtering the dataset for movies
movies_data = netflix_data[netflix_data['type'] == 'Movie']
```

```python
# Creating a DataFrame for movies under 60 minutes
short_movies = movies_data[movies_data['duration'] < 60]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Grouping by release year and calculating the mean duration for each year
yearly_duration = movies_data.groupby('release_year')['duration'].mean().reset_index()

# Plotting the trend
plt.figure(figsize=(12, 6))
sns.lineplot(data=yearly_duration, x='release_year', y='duration')
plt.title('Average Movie Duration on Netflix by Release Year')
plt.xlabel('Release Year')
plt.ylabel('Average Duration (Minutes)')
plt.grid(True)
plt.show()
```
